#### Remove sequence description from the fasta file to prepare for ESM embeddings

```python
from Bio import SeqIO

def edit_fasta_ids(input_fasta, output_fasta):
    with open(input_fasta, "r") as infile, open(output_fasta, "w") as outfile:
        for record in SeqIO.parse(infile, "fasta"):
            record.id = record.id.split(' ')[0]
            record.description = ''  # Ensure the full description is also updated
            SeqIO.write(record, outfile, "fasta")

# Example usage
input_fasta = '../data/processed/fasta/cyclase/PF05147_hits_200_500aa_U.fasta'  # replace with your input FASTA file path
output_fasta = '../data/processed/fasta/cyclase/PF05147_hits_200_500aa_U_id.fasta'  # replace with your desired output FASTA file path
edit_fasta_ids(input_fasta, output_fasta)
```

#### Load embeddings and save all embeddings to a pkl file

##### ESM-1b

```python
import esm
import pickle
import torch

# Path to esm1b embeddings
FASTA_PATH = "../data/processed/fasta/cyclase/PF05147_hits_200_500aa_U_id.fasta" # Path to fasta
EMB_PATH = "../models/PF05147_esm1b" # Path to directory of embeddings 
EMB_LAYER = 33

key = []
Xs = []
for header, _seq in esm.data.read_fasta(FASTA_PATH):
    id = header.split('/')[0]
    des = header.split('/')[-1]
    key.append(id[0:14])
    fn = f'{EMB_PATH}/{id}/{des}.pt'
    embs = torch.load(fn)
    embs['mean_representations'][EMB_LAYER]
    Xs.append(embs['mean_representations'][EMB_LAYER])
Xs = torch.stack(Xs, dim=0).numpy()

# Prepare a key to embedding dictionary
key2mu = {}
for n in range(len(key)):
    key2mu[key[n]] = Xs[n]

# Save key2mu to a .pkl file
with open("../models/ESM1b_key2mu.pkl", "wb") as f:
    pickle.dump(key2mu, f)
```
